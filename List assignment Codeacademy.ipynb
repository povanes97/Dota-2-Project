```python
# creating a list
paintings=['The Two Fridays','My Dress Hangs Here', 'Tree of Hope', 'Self Portrait With Monkeys']
dates=[1939, 1933, 1946, 1940]
painting=list(zip(paintings,dates))
print(list(painting))
```

```python
#Appending list
painting.append(['The Broken Column', 1944])
painting.append(['The Wounded Deer', 1946])
painting.append(['Me and My Doll', 1937])
print(painting)
```

```python
#length of list
print(len(painting))
audio_tour_number=list(range(1,len(painting)+1))
print(audio_tour_number)
master_list=list(zip(painting,audio_tour_number))
print(master_list)
```
