```python
#Comparing lane and jungle creep gold 

import statistics
import matplotlib.pyplot as plt

#neutral_camp_gold_ranges = {
 #   "Small": (56, 76),
  #  "Medium": (75, 93),
   # "Hard": (96, 107),
    #"Ancient": (131, 180)
#}
lane_creep_gold=[38,36,39,50,65]
ancient_creep_gold=[50,65,55]
hard_creep_gold=[30,35,33]
medium_creep_gold=[20,30,25]
small_creep_gold=[10,9,8,8,3]
```

```python
#calculate the sum lane creep


total_lane_creep_gold=0
for i in lane_creep_gold:
    total_lane_creep_gold+=i

print(total_lane_creep_gold)

#calculate the mean and variance/std of lane creep gold
mean_lane_creep_gold=statistics.mean(lane_creep_gold)
variance_lane_creep_gold=statistics.variance(lane_creep_gold)
standard_lane_creep_gold=statistics.stdev(lane_creep_gold)
print(mean_lane_creep_gold)
print(round(variance_lane_creep_gold,2))
print(round(standard_lane_creep_gold,2))

#Plot graph for lane creep gold
```

```python
#calculate the sum ancient creep

total_ancient_creep_gold=0
for i in ancient_creep_gold:
    total_ancient_creep_gold+=i

print(total_ancient_creep_gold)

#calculate the mean and variance/std of ancient creep gold

mean_ancient_creep_gold=statistics.mean(ancient_creep_gold)
variance_ancient_creep_gold=statistics.variance(ancient_creep_gold)
standard_ancient_creep_gold=statistics.stdev(ancient_creep_gold)
print(mean_ancient_creep_gold)
print(round(variance_ancient_creep_gold,2))
print(round(standard_ancient_creep_gold,2))
```

```python
#calculate the sum of hard camp creep

total_hard_creep_gold=0
for i in hard_creep_gold:
    total_hard_creep_gold+=i



print(total_hard_creep_gold)

#calculate the mean and variance/std of hard creep gold

mean_hard_creep_gold=statistics.mean(hard_creep_gold)
variance_hard_creep_gold=statistics.variance(hard_creep_gold)
standard_hard_creep_gold=statistics.stdev(hard_creep_gold)
print(round(mean_hard_creep_gold,2))
print(round(variance_hard_creep_gold,2))
print(round(standard_hard_creep_gold,2))
```

```python
#calculate the sum of medium camp creep

total_medium_creep_gold=0
for i in medium_creep_gold:
    total_medium_creep_gold+=i



print(total_medium_creep_gold)

#calculate the mean and variance/std of medium creep gold

mean_medium_creep_gold=statistics.mean(medium_creep_gold)
variance_medium_creep_gold=statistics.variance(medium_creep_gold)
standard_medium_creep_gold=statistics.stdev(medium_creep_gold)
print(round(mean_medium_creep_gold,2))
print(round(variance_medium_creep_gold,2))
print(round(standard_medium_creep_gold,2))
```

```python
#calculate the sum of small camp creep

total_small_creep_gold=0
for i in small_creep_gold:
    total_small_creep_gold+=i



print(total_small_creep_gold)

#calculate the mean and variance/std of small creep gold

mean_small_creep_gold=statistics.mean(small_creep_gold)
variance_small_creep_gold=statistics.variance(small_creep_gold)
standard_small_creep_gold=statistics.stdev(small_creep_gold)
print(round(mean_small_creep_gold,2))
print(round(variance_small_creep_gold,2))
print(round(standard_small_creep_gold,2))
```

```python
#Plot graph between lane creep,ancient creep,hard creep,medium creep,small creep


#sum
labels=['Lane','Ancient','Hard','Medium','Small']
total_values=[total_lane_creep_gold,total_ancient_creep_gold,total_hard_creep_gold,total_medium_creep_gold,total_small_creep_gold]

plt.figure(figsize=(6,4))
plt.bar(labels,total_values, color=['teal', 'orange','magenta','cyan','pink'])
plt.ylabel('Total Gold')
plt.title('Total between Jungle Creeps and Lane Creeps')
plt.show()

#mean
labels=['Lane','Ancient','Hard','Medium','Small']
mean_values=[mean_lane_creep_gold,mean_ancient_creep_gold,mean_hard_creep_gold,mean_medium_creep_gold,mean_small_creep_gold]
plt.figure(figsize=(6,4))
plt.bar(labels,mean_values, color=['green', 'darkred','navy','teal','brown'])
plt.ylabel('Average Gold')
plt.title('Average Gold: Lane Creeps vs Jungle Creeps')
plt.show()


#std
labels=['Lane','Ancient','Hard','Medium','Small']
std_values=[standard_lane_creep_gold,standard_ancient_creep_gold,standard_hard_creep_gold,standard_medium_creep_gold,standard_small_creep_gold]
plt.figure(figsize=(6,4))
plt.bar(labels,std_values, color=['olive', 'lime','grey','purple','orange'])
plt.ylabel('Standard deviation Gold')
plt.title('Standard deviation Gold: Lane Creeps vs Jungle Creeps')
plt.show()

```
